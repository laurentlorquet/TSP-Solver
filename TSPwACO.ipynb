In this project, we implemented an ant colony optimization algorithm in Python to approximate solutions for the Traveling Salesman Problem (TSP). The TSP is a classic optimization problem where the goal is to find the shortest route visiting each location exactly once and returning to the starting location. It is an NP-hard problem, meaning there is no known efficient algorithm to find optimal solutions as the number of cities grows.

Since the TSP cannot be solved optimally in polynomial time, we used a metaheuristic search algorithm based on the behavior of ant colonies to find good approximate solutions. In ACO, artificial ants iteratively construct solutions guided by pheromone trails and heuristic information. Although ACO does not guarantee optimality, it can provide high-quality solutions for large TSP instances fairly quickly.

In this report, we explain our Python implementation of an ACO algorithm for the TSP. We describe the overall approach, including how pheromone trails are initialized and updated, how ants build solutions probabilistically at each iteration, and how the colony converges over time. Key ACO components like evaporation, intensification, and diversification are discussed. Parameters chosen for factors like pheromone influence are explained.

Results on sample TSP instances are presented to demonstrate the algorithm's ability to converge to strong solutions. While not perfect, our ACO algorithm provides a reasonably fast technique to find near-optimal routes for traveling salesman problems. The ant colony's search process collaboratively learns over time to produce shorter tours.

#Basic Functionality
In crafting our approach, we followed a set of heuristics, focusing on finding solid solutions rather than striving for the ultimate optimal outcome. We harnessed the power of 'Ant Colony Optimization' (ACO), a metaheuristic search technique inspired by ant colony behavior. The ACO algorithm operates with several parameters: the number of iterations, the number of ants, pheromone evaporation rate, a list of cities, and a stagnation limit for iterations. At each iteration, artificial ants construct complete tours by probabilistically choosing the next location based on pheromone levels on graph edges. Over iterations, the ants collectively modify the pheromone trails to focus on high quality solutions. By balancing exploration and exploitation, the ant colony is able to converge on near-optimal tours for the TSP.

Our ACO algorithm utilizes several key functions:

*   An initialization function constructs the graph nodes representing cities, initializes pheromone trails to a small constant value on all edges, and sets parameters like number of ants, evaporation rate rho, etc.
*   A route construction function handles how each ant probabilistically builds a solution tour using pseudo random proportional rule based on edge pheromone and heuristic info. It keeps track of the visited cities.
*   A pheromone update function manages the pheromone deposit and evaporation on edges after each iteration. The best ant deposits more pheromone while trails also evaporate.
*   A stagnation check function tests for lack of improvement over a threshold of iterations and triggers termination.

The algorithm manipulates several key variables:

*   A pheromone matrix stores current pheromone levels on graph edges
*   Heuristic info matrix represents desired edge weights based on inter-location distances
*   Best tour cost variable tracks the shortest route found so far
*   We represent solutions as ordered lists of location indices, like [0, 3, 1, 2]. Fitness is the total tour distance cost calculated from the distance matrix.

The algorithm runs for a defined max iterations. At each iteration t, n ants construct tours probabilistically then pheromone is updated. Trail levels converge toward shorter tours over time.

Stagnation detection uses a counter s of iterations since last improvement. If s exceeds the max stagnation s_max, the algorithm terminates.

The ACO approach combines stochastic solution construction with biased pheromone learning to find high-quality solutions for TSP without getting trapped in local optima. It balances exploration of the search space with exploitation of accrued knowledge. The ant colony's collective intelligence emerges through simple local agent behaviors and pheromone feedback.

```python
# Group Members:
# Laurent Lorquet (Architect)
# Sebastian Rojas (Developer)
# Cedric Winter (Reporter)
# Professor: Dr. Oge Marques
import numpy as np
import random
import matplotlib.pyplot as plt

# Plot epochs on x-axis and shortest path distances on y-axis
def visualize_epoch_vs_shortest_path(epochs, shortest_distances):
    plt.plot(epochs, shortest_distances, color = "green")
    plt.title("\nBest Solution Distance Over Course of Search")
    plt.xlabel("Epoch(s)") # X axis
    plt.ylabel("Total distance (km) for TSP") # Y axis
    plt.show()

# Graph a scattered line plot of the best tour
def plot_optimal_path_through_cities(x, y):
    plt.title("\nOptimal Path Through Cities Found")
    plt.xlabel("X-Coordinate of Location")
    plt.ylabel("Y-Coordinate of Location")
    plt.plot(x, y, marker = "*", mec = "red", mfc = 'red', ms = 8, color = "green")
    plt.show()

# Class to represent Locations with a name and x, y coordinates for positioning
class Location:
    def __init__(self, title, x_coord, y_coord):
        self.title = title
        self.x_coord = x_coord
        self.y_coord = y_coord

    # Compute the distance between the current location and a given location
    def distance(self, location):
        return ((self.x_coord - location.x_coord)**2 + (self.y_coord - location.y_coord)**2)**0.5

    # Provides a string representation for the location object, which is its name
    def __repr__(self):
        return self.name

# Class for a tour made up of multiple cities
class Tour:
    def __init__(self, cities):
        # The tour starts and ends at the same location
        self.cities = cities + [cities[0]]

     # Compute the total distance of the tour
    def total_distance(self):
        return sum(self.cities[i].distance(self.cities[i + 1]) for i in range(len(self.cities) - 1))

    def __repr__(self):
        return ' -> '.join(location.name for location in self.cities)

# class to implement the aco
class AntColonyOptimization:
    def __init__(self, cities, n_ants, n_iterations, decay=0.99):
        self.cities = cities
        self.distances = self.calculate_distances()
        self.pheromone = np.ones(self.distances.shape) / len(cities)  # Initial pheromone concentration
        self.all_indxs = range(len(cities))  # Indexes of all the cities
        self.n_ants = n_ants
        self.n_iterations = n_iterations
        self.decay = decay  # Pheromone decay factor

    # Function calculates the distance between each pair of cities and stores it in a 2D numpy array
    def calculate_distances(self):
        n_cities = len(self.cities)
        distances = np.zeros((n_cities, n_cities))  # Initialize the 2D array with zeros
        for i in range(n_cities):
            for j in range(i+1, n_cities):
                distances[i, j] = self.cities[i].distance(self.cities[j])
                distances[j, i] = distances[i, j]  # distance must be symmetric
        return distances

    #Function runs the ACO algorithm for a given number of iterations
    def run(self):
        # Initial shortest path is infinitely long
        shortest_path = ("placeholder", np.inf)
        # Epoch counter
        epoch = 1
        x_arr_line_plot = []
        y_arr_line_plot = []
        # Iterate for a given number of times
        for i in range(self.n_iterations):
            # Generate all paths for the current iteration
            all_paths = self.gen_all_paths()
            # Spread pheromone along the paths
            self.propagate_pheromone_trails(all_paths)
            # Find the shortest path in the current iteration
            min_path = min(all_paths, key=lambda x: x[1])

            print(("Epoch " + str(epoch)) + " | " + "Minimum Total Distance: " + str(shortest_path[1]))
            epoch += 1
            x_arr_line_plot.append(epoch)
            y_arr_line_plot.append(shortest_path[1])
            if min_path[1] < shortest_path[1]: # If shorter than current shortest do your thing and update and decay
                shortest_path = min_path
            self.pheromone *= self.decay

        # Graph the min. distance at each iteration
        visualize_epoch_vs_shortest_path(x_arr_line_plot, y_arr_line_plot)

        return shortest_path

    # Function spreads the pheromone on the shortest path found in the current iteration
    def propagate_pheromone_trails(self, all_ant_paths):
        min_path = min(all_ant_paths, key=lambda x: x[1])  # Shortest path in the current iteration
        # For each move in the shortest path, increase the amount of pheromone
        for move in min_path[0]:
            self.pheromone[move] += 1.0 / self.distances[move]

    # Function generates all paths for all ants in the current iteration
    def gen_all_paths(self):
        all_paths = []  # List to store all paths
        for i in range(self.n_ants):
            path = self.gen_path(random.randint(0,24))
            all_paths.append((path, self.gen_path_dist(path)))
        return all_paths

    # Function generates a single path for one ant, starting from a given location
    def gen_path(self, start):
        path = []
        visited = set()  # Set to store visited cities
        visited.add(start)  # Add the start location to the visited set
        prev = start  # The previous location is the start location

        for i in range(len(self.distances) - 1):
            # Choose the next location to visit
            move = self.pick_move(self.pheromone[prev], self.distances[prev], visited)
            # Add the move to the path
            path.append((prev, move))
            prev = move  # previous location becomes the location just visited
            visited.add(move)  # Add the visited location to the set
        #  path ends with a move to the start location
        path.append((prev, start))

        return path

    # Function chooses the next location to move to
    def pick_move(self, pheromone, dist, visited):
        pheromone = np.copy(pheromone)  # Copy of the pheromone list
        # Set the pheromone value to 0 for the cities that have already been visited
        pheromone[list(visited)] = 0
        dist[dist == 0] = 1e-10  # avoid division by zero
        # Probability of moving to each location
        row = pheromone / dist
        norm_row = row / row.sum()
        # Choosing location to move to based on the made up probability I created
        move = np.random.choice(self.all_indxs, 1, p=norm_row)[0]
        return move

    # This function calculates the total distance of a given path
    def gen_path_dist(self, path):
        return sum([self.distances[i, j] for i, j in path])

# List of location names
location_names = ["Metropolis", "Capital City", "Harborton", "Creekdale", "Riverbend",
              "Greenland", "Gardenside", "Goldcoast", "Mountainview", "Aerie Peak",
              "Marina Bay", "Redcliff", "Desert Oasis", "Dragon Isles", "Blackrock Isle",
              "Old Haven", "Thunder Bay", "Seabreeze", "Whitesail", "Ironforge", "Emerald Cove",
              "Silvershore", "Stormcoast", "Farland", "Tradeford"]

# Generate list of location objects with random coordinates
location_list = [Location(name, random.uniform(-200.0, 200.0), random.uniform(-200.0, 200.0)) for name in location_names]

print("---------------------------------------------------------------------")
print(("Parameters for Ant Colony Optimization", "bold"))
print("---------------------------------------------------------------------")

# Print the names of the locations in the best tour and its total distance
print(("Number of cities in the tour:"), len(location_names))

# Prompt for number of iterations
#n_iterations = int(input("Enter the number of iterations (default = 200): "))
print("\n")

# Initialize instance of the AntColonyOptimization class
aco = AntColonyOptimization(location_list, n_ants=25, n_iterations=200)

# Run ACO algorithm and store the shortest path found
shortest_path = aco.run()

# Print best tour found out of all the paths and total distance
print(("\nBest Tour:", "bold"))
i = 1
x_arr_for_scatter = []
y_arr_for_scatter = []
for move in shortest_path[0]:
    print(f"{i}. {(location_list[move[0]].title)} | Coodinates = ({location_list[move[0]].x_coord}, {location_list[move[0]].y_coord})")
    i += 1
    x_arr_for_scatter.append(location_list[move[0]].x_coord)
    y_arr_for_scatter.append(location_list[move[0]].y_coord)
print(("\nTotal distance:", "bold"), shortest_path[1])
plot_optimal_path_through_cities(x_arr_for_scatter, y_arr_for_scatter)
```

```python
# Group Members:
# Laurent Lorquet (Architect)
# Sebastian Rojas (Developer)
# Cedric Winter (Reporter)
# Professor: Dr. Oge Marques
import numpy as np
import random
import matplotlib.pyplot as plt

# Plot epochs on x-axis and shortest path distances on y-axis
def visualize_epoch_vs_shortest_path(epochs, shortest_distances):
    plt.plot(epochs, shortest_distances, color = "green")
    plt.title("\nBest Solution Distance Over Course of Search")
    plt.xlabel("Epoch(s)") # X axis
    plt.ylabel("Total distance (km) for TSP") # Y axis
    plt.show()

# Graph a scattered line plot of the best tour
def plot_optimal_path_through_cities(x, y):
    plt.title("\nOptimal Path Through Cities Found")
    plt.xlabel("X-Coordinate of Location")
    plt.ylabel("Y-Coordinate of Location")
    plt.plot(x, y, marker = "*", mec = "red", mfc = 'red', ms = 8, color = "green")
    plt.show()

# Class to represent Locations with a name and x, y coordinates for positioning
class Location:
    def __init__(self, title, x_coord, y_coord):
        self.title = title
        self.x_coord = x_coord
        self.y_coord = y_coord

    # Compute the distance between the current location and a given location
    def distance(self, location):
        return ((self.x_coord - location.x_coord)**2 + (self.y_coord - location.y_coord)**2)**0.5

    # Provides a string representation for the location object, which is its name
    def __repr__(self):
        return self.name

# Class for a tour made up of multiple cities
class Tour:
    def __init__(self, cities):
        # The tour starts and ends at the same location
        self.cities = cities + [cities[0]]

     # Compute the total distance of the tour
    def total_distance(self):
        return sum(self.cities[i].distance(self.cities[i + 1]) for i in range(len(self.cities) - 1))

    def __repr__(self):
        return ' -> '.join(location.name for location in self.cities)

# class to implement the aco
class AntColonyOptimization:
    def __init__(self, cities, n_ants, n_iterations, decay=0.99):
        self.cities = cities
        self.distances = self.calculate_distances()
        self.pheromone = np.ones(self.distances.shape) / len(cities)  # Initial pheromone concentration
        self.all_indxs = range(len(cities))  # Indexes of all the cities
        self.n_ants = n_ants
        self.n_iterations = n_iterations
        self.decay = decay  # Pheromone decay factor

    # Function calculates the distance between each pair of cities and stores it in a 2D numpy array
    def calculate_distances(self):
        n_cities = len(self.cities)
        distances = np.zeros((n_cities, n_cities))  # Initialize the 2D array with zeros
        for i in range(n_cities):
            for j in range(i+1, n_cities):
                distances[i, j] = self.cities[i].distance(self.cities[j])
                distances[j, i] = distances[i, j]  # distance must be symmetric
        return distances

    #Function runs the ACO algorithm for a given number of iterations
    def run(self):
        # Initial shortest path is infinitely long
        shortest_path = ("placeholder", np.inf)
        # Epoch counter
        epoch = 1
        x_arr_line_plot = []
        y_arr_line_plot = []
        # Iterate for a given number of times
        for i in range(self.n_iterations):
            # Generate all paths for the current iteration
            all_paths = self.gen_all_paths()
            # Spread pheromone along the paths
            self.propagate_pheromone_trails(all_paths)
            # Find the shortest path in the current iteration
            min_path = min(all_paths, key=lambda x: x[1])

            print(("Epoch " + str(epoch)) + " | " + "Minimum Total Distance: " + str(shortest_path[1]))
            epoch += 1
            x_arr_line_plot.append(epoch)
            y_arr_line_plot.append(shortest_path[1])
            if min_path[1] < shortest_path[1]: # If shorter than current shortest do your thing and update and decay
                shortest_path = min_path
            self.pheromone *= self.decay

        # Graph the min. distance at each iteration
        visualize_epoch_vs_shortest_path(x_arr_line_plot, y_arr_line_plot)

        return shortest_path

    # Function spreads the pheromone on the shortest path found in the current iteration
    def propagate_pheromone_trails(self, all_ant_paths):
        min_path = min(all_ant_paths, key=lambda x: x[1])  # Shortest path in the current iteration
        # For each move in the shortest path, increase the amount of pheromone
        for move in min_path[0]:
            self.pheromone[move] += 1.0 / self.distances[move]

    # Function generates all paths for all ants in the current iteration
    def gen_all_paths(self):
        all_paths = []  # List to store all paths
        for i in range(self.n_ants):
            path = self.gen_path(random.randint(0,24))
            all_paths.append((path, self.gen_path_dist(path)))
        return all_paths

    # Function generates a single path for one ant, starting from a given location
    def gen_path(self, start):
        path = []
        visited = set()  # Set to store visited cities
        visited.add(start)  # Add the start location to the visited set
        prev = start  # The previous location is the start location

        for i in range(len(self.distances) - 1):
            # Choose the next location to visit
            move = self.pick_move(self.pheromone[prev], self.distances[prev], visited)
            # Add the move to the path
            path.append((prev, move))
            prev = move  # previous location becomes the location just visited
            visited.add(move)  # Add the visited location to the set
        #  path ends with a move to the start location
        path.append((prev, start))

        return path

    # Function chooses the next location to move to
    def pick_move(self, pheromone, dist, visited):
        pheromone = np.copy(pheromone)  # Copy of the pheromone list
        # Set the pheromone value to 0 for the cities that have already been visited
        pheromone[list(visited)] = 0
        dist[dist == 0] = 1e-10  # avoid division by zero
        # Probability of moving to each location
        row = pheromone / dist
        norm_row = row / row.sum()
        # Choosing location to move to based on the made up probability I created
        move = np.random.choice(self.all_indxs, 1, p=norm_row)[0]
        return move

    # This function calculates the total distance of a given path
    def gen_path_dist(self, path):
        return sum([self.distances[i, j] for i, j in path])

# List of location names
location_names = ["Metropolis", "Capital City", "Harborton", "Creekdale", "Riverbend",
              "Greenland", "Gardenside", "Goldcoast", "Mountainview", "Aerie Peak",
              "Marina Bay", "Redcliff", "Desert Oasis", "Dragon Isles", "Blackrock Isle",
              "Old Haven", "Thunder Bay", "Seabreeze", "Whitesail", "Ironforge", "Emerald Cove",
              "Silvershore", "Stormcoast", "Farland", "Tradeford"]

# Generate list of location objects with random coordinates
location_list = [Location(name, random.uniform(-200.0, 200.0), random.uniform(-200.0, 200.0)) for name in location_names]

print("---------------------------------------------------------------------")
print(("Parameters for Ant Colony Optimization", "bold"))
print("---------------------------------------------------------------------")

# Print the names of the locations in the best tour and its total distance
print(("Number of cities in the tour:"), len(location_names))

# Prompt for number of iterations
#n_iterations = int(input("Enter the number of iterations (default = 200): "))
print("\n")

# Initialize instance of the AntColonyOptimization class
aco = AntColonyOptimization(location_list, n_ants=25, n_iterations=500)

# Run ACO algorithm and store the shortest path found
shortest_path = aco.run()

# Print best tour found out of all the paths and total distance
print(("\nBest Tour:", "bold"))
i = 1
x_arr_for_scatter = []
y_arr_for_scatter = []
for move in shortest_path[0]:
    print(f"{i}. {(location_list[move[0]].title)} | Coodinates = ({location_list[move[0]].x_coord}, {location_list[move[0]].y_coord})")
    i += 1
    x_arr_for_scatter.append(location_list[move[0]].x_coord)
    y_arr_for_scatter.append(location_list[move[0]].y_coord)
print(("\nTotal distance:", "bold"), shortest_path[1])
plot_optimal_path_through_cities(x_arr_for_scatter, y_arr_for_scatter)
```
